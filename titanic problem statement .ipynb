# 2) Find the Third class travled people and find the count of teenage and adult age and old age.

```python
dataset
```

```python
data1=dataset.iloc[:,[8,3]]
data1
```

```python
data1["age"].value_counts()
```

```python
AGE=[]
for x in data1["age"]:
    if x>0 and x<=15:
        AGE.append("child")
    elif x>15 and x<=30:
        AGE.append("teenage")
    elif x>30 and x<=50:
        AGE.append("adult")
    else:
        AGE.append("old age")
print(AGE)  
```

```python
data1.insert(0,"AGE",AGE)
```

```python
del data1["age"]
```

```python
data1
```

```python
data2=data1.loc[(data1["class"]=="Third")&(data1["AGE"]!="child")]
data2
```
